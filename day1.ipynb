```python
import numpy as np
```

```python
literals = [ "one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]
map = {literals[i-1]: str(i) for i in range(1,10)}
```

```python
combined = []
with open('day1.txt') as f:
    for i, line in enumerate(f):
        numbers = []
        intm = ""
        for char in line:
            if char.isnumeric():
                numbers.append(char)
                intm = ""
            else:
                intm += char
                for i, lit in enumerate(literals):
                    if lit in intm:
                        numbers.append(str(i+1))
                        intm = intm[-1]
                        break
                    
            
        combined.append(int(numbers[0] + numbers[-1]))
        print(f"{i+1}: {numbers} -> {int(numbers[0] + numbers[-1])}")

```
